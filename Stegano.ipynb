```python
#importing necessary libraries
from PIL import Image
import os
import shutil

class Stegano:

    def __init__(self,Iname):

        self.name = Iname

        #copying the image from Desktop to working Directory
        shutil.copyfile('C:\\Users\MOHAN\Desktop\\' + self.name,
                        'C:\\Users\MOHAN\Desktop\\Steganography\\' + self.name)

        # returns the size of the file in Bytes only when file is present in working directory
        self.size = os.path.getsize(Iname)
        self.im = Image.open(Iname)

        #storing the width and height of image
        self.x, self.y = self.im.size


    def encode(self,message):
        '''function to encode a message inside the image
        message: string to hide inside the image'''

        #creating a copy of the image
        self.im_c = self.im.copy()

        #adding **** at the end of the message
        msg =  message +'****'

        #converting message to binary
        bin_msg = "".join([format(ord(i),"08b") for i in msg])

        #checking if size of image is less than the size of message
        if self.size < len(bin_msg):
            print("INSUFFICIENT SIZE of image")
            exit(0)

        #initialising i with 0
        i = 0

        #loops: first to moving along x_axis then moving along y-axis
        for j in range(self.y):
            for k in range(self.x):
                #extracting r,g,b pixel values from the image at the specified co-ordinate
                r, g, b = self.im_c.getpixel((k, j))

                if i < len(bin_msg):
                    #if pixel value is even and bin value of msg is 0 OR
                    #pixel value is odd and bin value is 1 then change it otherwise don't

                    if(r % 2 == 0 and bin_msg[i] == '0') or (r % 2 != 0 and bin_msg[i] == '1'):
                        r = r - 1
                        i += 1
                    else:
                        i+=1
                if i < len(bin_msg):
                    if (g % 2 == 0 and bin_msg[i] == '0') or (g % 2 != 0 and bin_msg[i] == '1'):
                        g = g - 1
                        i += 1
                    else:
                        i+=1
                if i < len(bin_msg):
                    if (b % 2 == 0 and bin_msg[i] == '0') or (b % 2 != 0 and bin_msg[i] == '1'):
                        b = b - 1
                        i += 1
                    else:
                        i+=1
                self.im_c.putpixel((k,j),(r,g,b))

        #splitting the image name
        fname = (self.name.split('.'))[0]

        #specifying path of desktop and adding ENCODED to image name
        new_im = 'C:\\Users\MOHAN\Desktop\Encoded' + fname + '.png'

        #saving the encoded image
        self.im_c.save(new_im)

        return 'Image Encoded'


    def decode(self):
        '''Function to decode the message from the image of png extension'''
        result =''
        bin_msg = ''
        k=0
        for i in list(self.im.getdata()):
            for j in i:
                if(j%2 == 0):
                    bin_msg+='1'
                else:
                    bin_msg+='0'
                
                if(len(bin_msg)%8 == 0):
                    result+=chr(int(bin_msg[k:k+8],2))

                #checking if at the end of decoded msg there are **** which we added during encoding
                    if result[-4:] == '****':
                        return ('Hidden Message: {}'.format(result[:-4]))
                    k+=8

if __name__ == '__main__':

    print('--------IMAGE STEGANOGRAPHY----------')
    print('Note: Image should be on Desktop')
    print("1.Encode")
    print("2.Decode")
    ch = input('Enter Your Choice\n')
    if ch == '1':
        name = input('Enter Image Name with Extension\n')
        s = Stegano(name)
        msg = input("Enter the msg to hide\n")
        print(s.encode(msg))
    elif ch == '2':
        print('Image should have png extension')
        name = input('Enter Image with Extension \n')
        if name[-3:]!='png':
            print('Cannot Decode the Image, Image should have png extension')
        else:
            s = Stegano(name)
            print(s.decode())
    else:
        print('Wrong Choice')
```
